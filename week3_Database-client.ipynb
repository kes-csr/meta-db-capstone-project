```python
show_tables_query = "SHOW tables" 
cursor.execute(show_tables_query)
```

```python
results_tables = cursor.fetchall()

for table in results_tables:
    print(table)
```

```python
join_tables_query = """
SELECT 
CONCAT(Customers.FirstName,' ',Customers.LastName) As 'Full Name',
Customers.ContactNumber AS 'Phone Number',
Customers.Email AS 'E-mail',
Orders.TotalCost AS 'Bill Amount $'
From Orders INNER JOIN Customers
ON Orders.CustomerID = Customers.CustomerID
WHERE Orders.TotalCost > 60;
"""
```

```python
cursor.execute(join_tables_query)
results_join_tables =cursor.fetchall()
columns_join_tables = cursor.column_names
print(cursor.column_names)
for row in results_join_tables:
    print(row)
```
